## Exercise 1
#### A)
The only way to guarantee that all odd squares are =1 mod(8) is by using generic terms as its possible to see above:

Given any odd number, it can be represented as \(2n + 1\) where \(n\) is an integer.

Now, let's square it:
<div style="margin-left: 20px;">
    <p><code>(2n + 1)<sup>2</sup> = 4n<sup>2</sup> + 4n + 1</code></p>
    <p><code>= 4n(n + 1) + 1</code></p>
</div>

Now, either \(n\) or \(n + 1\) is even since one will be odd and the other even. Thus, \(n(n + 1)\) is even.

So, the expression can be written as:
<div style="margin-left: 20px;">
    <p><code>4n(n + 1) + 1 = 8m + 1</code></p>
    <p>where <code>2m = n(n + 1)</code> is an integer.</p>
</div>

Now, let's find the modulus when divided by 8:
<div style="margin-left: 20px;">
    <p><code>(8m + 1) mod 8 = 1</code></p>
</div>

Hence, its true that all odd squares are congruent to 1 modulo 8.


```python
from sympy import symbols, Eq, Mod

length = 1000 # You can increase the range for more testing
n = symbols('n', integer=True)

odd_num = 2*n + 1

square = odd_num**2 # 4**2 + 4n + 1

mod_result = Mod(square, 8)

for i in range(length): 
    result = mod_result.subs(n, i)
    assert result == 1

print("All tested odd squares are congruent to 1 modulo 8.")
```

```python
from sympy import symbols, Eq, Mod

length = 1000 # You can increase the range for more testing
n = symbols('n', integer=True)
alternator = False
even_num = 2*n

square = even_num**2 # 4*n**2

mod_result = Mod(square, 8)

for i in range(length): 
    result = mod_result.subs(n, i)
    assert result != 1
    if alternator == False:
        alternator = True
        assert result == 0
    else:
        alternator = False
        assert result == 4
        
    #assert result == 0 or result == 4

print("All tested even squares are not congruent to 1 modulo 8. All tested even squares are 0 or 4")
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1, 100, 400)
n = n.astype(int)


O_n = n
O_1 = np.ones_like(n)
O_logn = np.log2(n)


plt.figure(figsize=(10, 6))
plt.ylim(0, 50)  
plt.plot(n, O_n, label="O(n)")
plt.plot(n, O_1, label="O(1)")
plt.plot(n, O_logn, label="O(log n)")


plt.legend()
plt.title("Comparison of Complexity Classes")
plt.xlabel("n")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

As perceived from the above graph the best option would be a proofSize of O(1) which indicates that my proof won't grow in size given my input. That way I can store the proof safely on a blockchain for instance, where storage comes at a premium.
In second place a logarithmic proofSize would be also somewhat good given that my proof will be growing slower per the input
